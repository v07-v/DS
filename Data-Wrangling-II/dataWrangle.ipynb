```python
missing_values = ["Na","na"]
df = pd.read_csv('./StudentPerformance.csv', na_values = missing_values)
df
```

```python
ndf=df  # it just creating a view of dataframe and not altering the actual dataframe
ndf.fillna(0)
```

```python
m_v=df['math score'].mean()
df['math score'] = df['math score'].fillna(m_v)
df.head(10)
```

```python
me_v=df['math score'].median()
df['math score'] = df['math score'].fillna(me_v)
df.head(10)
```

```python
std_v=df['writing score'].std()
df['writing score'] = df['writing score'].fillna(std_v)
df.head(10)
```

```python
min_v=df['Placement Score'].min()
df['Placement Score']=df['Placement Score'].fillna(min_v)
df.head(10)
```

```python
max_v=df['Placement Score'].max()
df['Placement Score']=df['Placement Score'].fillna(max_v)
df.head(10)
```
